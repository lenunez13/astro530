```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter, NullFormatter,FormatStrFormatter
from PyAstronomy.pyasl import planck
from scipy import integrate
import matplotlib.ticker as ticker
plt.rcParams.update({'font.size': 14})

from astro530 import Planck
from astro530 import Integrate
from astro530 import Constants

G,c,h,k_B,sigma_sb = Constants()
```

```python
#...Initiate wavenumbers and temperatures
nu_tildes = np.linspace(0.1,12.,100)*1e6 #convert 1/microns to 1/m
T1,T2,T3 = 10000., 7000., 3000,

#...Plot Planck function versus wavenumber
fig, ax = plt.subplots()
ax.plot(nu_tildes/1e6,Planck(T1,nu_tilde=nu_tildes),label=r'$T=10000$ K',ls='-',c='k')
ax.plot(nu_tildes/1e6,Planck(T2,nu_tilde=nu_tildes),label=r'$T=7000$ K',ls='--',c='k')
ax.plot(nu_tildes/1e6,Planck(T3,nu_tilde=nu_tildes),label=r'$T=3000$ K',ls=':',c='k')
ax.set_xlabel(r'$\~{\nu}\,(\mathrm{\mu m^{-1}})$')
ax.set_ylabel(r'$B_{\nu}$')
ax.legend()
plt.savefig("HW2_plot1.pdf",bbox_inches='tight')

fig, ax = plt.subplots()
ax.plot(nu_tildes/1e6,Planck(T1,nu_tilde=nu_tildes),label=r'$T=10000$ K',ls='-',c='k')
ax.plot(nu_tildes/1e6,Planck(T2,nu_tilde=nu_tildes),label=r'$T=7000$ K',ls='--',c='k')
ax.plot(nu_tildes/1e6,Planck(T3,nu_tilde=nu_tildes),label=r'$T=3000$ K',ls=':',c='k')
ax.set_yscale('log')
ax.set_xlabel(r'$\~{\nu}\,(\mathrm{\mu m^{-1}})$')
ax.set_ylabel(r'$B_{\nu}$')
ax.legend()
plt.savefig("HW2_plot2.pdf",bbox_inches='tight')

fig, ax = plt.subplots()
ax.plot(nu_tildes/1e6,Planck(T1,nu_tilde=nu_tildes),label=r'$T=10000$ K',ls='-',c='k')
ax.plot(nu_tildes/1e6,Planck(T2,nu_tilde=nu_tildes),label=r'$T=7000$ K',ls='--',c='k')
ax.plot(nu_tildes/1e6,Planck(T3,nu_tilde=nu_tildes),label=r'$T=3000$ K',ls=':',c='k')
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlabel(r'$\log{\~{\nu}}$')
ax.set_ylabel(r'$B_{\nu}$')
ax.set_xscale('log')
ax.xaxis.set_major_formatter(ticker.LogFormatterExponent())
ax.legend()
plt.savefig("HW2_plot3.pdf",bbox_inches='tight')

```

```python
funct = lambda nu_tilde: Planck(T,nu_tilde=nu_tilde) # Planck function vs. wavenumber
nu_tilde_0 = 0.
nu_tilde_f = 12.*1e6
T = 7500.

experimental = Integrate(funct,nu_tilde_0,nu_tilde_f,n=1e3)*c # User-defined integral over all wavelengths
accepted = sigma_sb/np.pi*T**4. # analytical (accepted) result, for comparison
relative_error = (experimental-accepted)/accepted
print("The relative error is "+str(relative_error)+".")
```
