```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

```

```python
plt.rcParams.update({'font.size': 14})
x = np.linspace(0,5.5,100)
fig, ax = plt.subplots()
ax.plot(x,5*np.exp(-x),'-',c='k',label=r'$5e^{-x}$')
ax.plot(x,5-x,'--',c='k',label=r'$5-x$')
ax.set_xlabel(r'$x$',fontsize=14)
ax.axhline(y=0,alpha=0.25,color='k')
ax.axvline(x=0,alpha=0.25,color='k')
ax.legend(prop={'size': 14})
plt.savefig('HW1.png',bbox_inches='tight',dpi=400)
```
